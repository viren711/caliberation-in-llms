```python
from groq import Groq
from google.colab import userdata

# 1. Safely grab your hidden API key from Colab's Secrets
my_key = userdata.get('GROQ_API_KEY')

# 2. Connect to Groq
client = Groq(api_key=my_key)

# 3. Ask Groq a question using the fast Llama 3.3 model
print("Asking Groq a question... watch how fast this is!")
chat_completion = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        # The 'system' role gives the AI its personality/instructions
        {"role": "system", "content": "You are a helpful and energetic assistant."},
        # The 'user' role is your actual question
        {"role": "user", "content": "Explain Quantum Computing in exactly one simple sentence."}
    ]
)

# 4. Print the blazing-fast answer
print("\nGroq says:")
print(chat_completion.choices[0].message.content)
```

```python
from openai import OpenAI
from google.colab import userdata

# 1. Safely grab your hidden API key from Colab's Secrets
my_key = userdata.get('OPENROUTER_API_KEY')

# 2. Connect to OpenRouter (using the OpenAI library!)
client = OpenAI(
    base_url="https://openrouter.ai/api/v1", # This is the magic redirect link
    api_key=my_key,
)

# 3. Ask a question using OpenRouter's automatic free model selector
print("Asking OpenRouter's free models a question...")
completion = client.chat.completions.create(
    model="openrouter/free",
    messages=[
        {"role": "system", "content": "You are a helpful science teacher."},
        {"role": "user", "content": "Explain what a overfitting in ml."}
    ]
)

# 4. Print the answer
print("\nOpenRouter says:")
print(completion.choices[0].message.content)
```

```python
from google import genai
from google.colab import userdata

# 1. Safely grab your hidden API key from Colab's Secrets
my_key = userdata.get('GEMINI_API_KEY')

# 2. Connect to Gemini
client = genai.Client(api_key=my_key)

# 3. Read the CSV file into a Python string
file_path = "t20_wc_2026_results.csv"
with open(file_path, "r", encoding="utf-8") as file:
    csv_data = file.read()

# 4. NEW PROMPT: Tell Gemini exactly how to format the ranking
prompt = f"""
Here is the raw CSV data for the T20 World Cup 2026 matches:

{csv_data}

Based strictly on this data, please analyze the performance of all the teams.
Figure out who the ultimate winner is (if there is a final) or who has the most wins.
Then, give me a ranking of the top 10 teams using pair wise ranking .

Please format your response EXACTLY like this:
Rank 1 - [Team Name] - Reason: [Explain why, e.g., "They won the Final", or "They had 6 total wins"]
Rank 2 - [Team Name] - Reason: [Explain why]
Rank 3 - [Team Name] - Reason: [Explain why]
"""

# 5. Ask Gemini Flash to analyze the data
print("Analyzing CSV data and calculating rankings...")
response = client.models.generate_content(
    model="gemini-2.5-flash",
    contents=prompt
)

# 6. Print the analysis
print("\n--- Overall Team Rankings ---")
print(response.text)
```

```python
from groq import Groq
from google.colab import userdata

# 1. Safely grab your hidden API key from Colab's Secrets
my_key = userdata.get('GROQ_API_KEY')

# 2. Connect to Groq
client = Groq(api_key=my_key)

# 3. Read the CSV file into a Python string
file_path = "t20_wc_2026_results.csv"
with open(file_path, "r", encoding="utf-8") as file:
    csv_data = file.read()

# 4. Create your prompt instructions (just the text part)
user_prompt = f"""
Here is the raw CSV data for the T20 World Cup 2026 matches:

{csv_data}

Based strictly on this data, please analyze the performance of all the teams.
Figure out who the ultimate winner is (if there is a final) or who has the most wins.
Then, give me a ranking of the top 10 teams using pair wise ranking.

Please format your response EXACTLY like this:
Rank 1 - [Team Name] - Reason: [Explain why, e.g., "They won the Final", or "They had 6 total wins"]
Rank 2 - [Team Name] - Reason: [Explain why]
Rank 3 - [Team Name] - Reason: [Explain why]
"""

# 5. Ask Groq to analyze the data
print("Analyzing CSV data with Groq (Llama 3.3)... watch the speed!")
chat_completion = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        # Tell the AI how to behave
        {"role": "system", "content": "You are an expert sports data analyst. Follow formatting instructions perfectly."},
        # Give it the CSV data and your question
        {"role": "user", "content": user_prompt}
    ]
)

# 6. Print the analysis
print("\n--- Overall Team Rankings (via Groq) ---")
print(chat_completion.choices[0].message.content)
```
