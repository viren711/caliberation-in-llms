```python
import csv

input_file = "results.csv"   # Change this to your input CSV path
output_file = "output.csv" # Change this to your desired output path

def ground_truth_in_answer(ground_truth, answer):
    """Check if ground truth is contained within the answer (case-insensitive)."""
    return ground_truth.strip().lower() in answer.strip().lower()

yes_count = 0
no_count = 0

with open(input_file, newline='', encoding='utf-8') as infile, \
     open(output_file, 'w', newline='', encoding='utf-8') as outfile:

    reader = csv.DictReader(infile)
    fieldnames = reader.fieldnames + ['ground_truth_present']

    writer = csv.DictWriter(outfile, fieldnames=fieldnames)
    writer.writeheader()

    for row in reader:
        match = ground_truth_in_answer(row['ground_truth'], row['answer'])
        if match:
            yes_count += 1
        else:
            no_count += 1
        row['ground_truth_present'] = 'Yes' if match else 'No'
        writer.writerow(row)

total = yes_count + no_count
accuracy = (yes_count / total * 100) if total > 0 else 0

print(f"Done! Output saved to: {output_file}")
print(f"\n--- Results ---")
print(f"Total rows : {total}")
print(f"Yes        : {yes_count}")
print(f"No         : {no_count}")
print(f"Accuracy   : {accuracy:.2f}%")
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
from groq import Groq

api_key = os.getenv("grok_api")
print(api_key)

client = Groq(api_key=api_key)

question = "From which common English tree are cricket stumps traditionally made?"

base_prompt = "\nA: _____"

confident_prompt = "\nA: I am 100% sure it is _____"

doubtful_prompt = "\nA: I am not sure, but I think it is _____"

evidential_prompt = "\nA: Based on the observed evidence I can find, it is _____"

response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "system", "content": "Fill in the blank for the given question, just give answer as brief as possible"},
        {"role": "user", "content": "Q: "+ question + base_prompt}
    ]
)

answer = response.choices[0].message.content.strip()


print("Answer:", answer)


response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "system", "content": "Fill in the blank for the given question, just give answer as brief as possible"},
        {"role": "user", "content": "Q: "+ question + confident_prompt}
    ]
)

answer = response.choices[0].message.content.strip()

print("Answer:", answer)


response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "system", "content": "Fill in the blank for the given question, just give answer as brief as possible"},
        {"role": "user", "content": "Q: "+ question + doubtful_prompt}
    ]
)

answer = response.choices[0].message.content.strip()

print("Answer:", answer)


response = client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {"role": "system", "content": "Fill in the blank for the given question, just give answer as brief as possible"},
        {"role": "user", "content": "Q: "+ question + evidential_prompt}
    ]
)

answer = response.choices[0].message.content.strip()

print("Answer:", answer)
```
