```python
import csv
import time
from google import genai
from google.colab import userdata

# 1. Load API key
api_key = userdata.get('Gemini_API')

# 2. Initialize Gemini client
client = genai.Client(api_key=api_key)

input_file = "part2_remaining.csv"
output_file = "results_gemini.csv"

# ✅ Create CSV file FIRST
with open(output_file, 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(["question", "answer"])

print("✅ CSV file created. Starting processing...\n")

with open(input_file, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)

    for row in reader:
        question = row['question']
        print(f"Processing: {question}")

        try:
            response = client.models.generate_content(
                model="gemini-2.5-flash",
                contents=f"Answer very briefly (max 5-10 words): {question}"
            )

            answer = response.text.strip()
            print("Answer:", answer)

        except Exception as e:
            print(f"\n❌ Error occurred: {e}")
            print("🛑 Stopping execution. Partial results saved in results.csv")
            break  # 🔥 STOP immediately

        # ✅ Append result instantly
        with open(output_file, 'a', newline='', encoding='utf-8') as f_out:
            writer = csv.writer(f_out)
            writer.writerow([question, answer])

        # ⚠️ IMPORTANT: Gemini rate limit protection
        time.sleep(12)   # 🔥 must be high to avoid 429

print("\n✅ Finished (or stopped due to error). Check results.csv")
```

```python
import csv
import time
from groq import Groq
from google.colab import userdata

# 1. Load API key
api_key = userdata.get('G_api')

# 2. Initialize Groq client
client = Groq(api_key=api_key)

input_file = "part2_remaining.csv"
output_file = "results_groq_from_10977.csv"

# ✅ Create CSV FIRST
with open(output_file, 'w', newline='', encoding='utf-8') as f:
    writer = csv.writer(f)
    writer.writerow(["question", "answer"])

print("✅ CSV created. Starting...\n")

with open(input_file, 'r', encoding='utf-8') as f:
    reader = csv.DictReader(f)

    for row in reader:
        question = row['question']
        print(f"Processing: {question}")

        try:
            chat_completion = client.chat.completions.create(
                model="llama-3.3-70b-versatile",
                messages=[
                    {"role": "system", "content": "Answer very briefly (max 5-10 words)."},
                    {"role": "user", "content": question}
                ]
            )

            answer = chat_completion.choices[0].message.content.strip()
            print("Answer:", answer)

        except Exception as e:
            print(f"\n❌ Error: {e}")
            print("🛑 Stopping. Partial results saved in results_groq.csv")
            break   # 🔥 STOP immediately on error

        # ✅ Append immediately (no data loss)
        with open(output_file, 'a', newline='', encoding='utf-8') as f_out:
            writer = csv.writer(f_out)
            writer.writerow([question, answer])

        time.sleep(0.5)

print("\n✅ Finished (or stopped early). Check results_groq.csv")
```

```python
from sentence_transformers import SentenceTransformer, util
import pandas as pd

model = SentenceTransformer('all-MiniLM-L6-v2')

df = pd.read_csv("final_eval_dataset.csv")

scores = []

for _, row in df.iterrows():
    emb1 = model.encode(str(row["ground_truth"]), convert_to_tensor=True)
    emb2 = model.encode(str(row["answer"]), convert_to_tensor=True)

    sim = util.cos_sim(emb1, emb2).item()
    scores.append(sim)

df["semantic_score"] = scores

accuracy = (df["semantic_score"] > 0.7).mean()

print("Semantic Accuracy:", accuracy)
```
